### This python notebook demonstrates implementing `dense optical flow` in OpenCV.

### `Lucas-Kanade` method computes optical flow for a sparse feature set (in our example, corners detected using `Shi-Tomasi` algorithm).

### OpenCV provides another algorithm to find the `dense optical flow`. It computes the optical flow for all the points in the frame. It is based on `Gunnar Farneback's` algorithm.

```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture("inputs/vtest.avi")
```

```python
ret, frame1 = cap.read()
prvs = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
hsv = np.zeros_like(frame1)
hsv[..., 1] = 255
```

```python
while (1):
    ret, frame2 = cap.read()
    if not ret:
        print('No frames grabbed')
        break

    next = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)
    flow = cv2.calcOpticalFlowFarneback(prvs, next, None, 0.5, 3, 15, 3, 5, 1.2, 0)
    mag, ang = cv2.cartToPolar(flow[..., 0], flow[..., 1])
    hsv[..., 0] = ang * 180 / np.pi / 2
    hsv[..., 2] = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)
    bgr = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

    cv2.imshow('frame2', bgr)

    k = cv2.waitKey(30) & 0xFF
    if k == ord('q'):
        cv2.imwrite('opticalfb.png', frame2)
        cv2.imwrite('opticalhsv.png', bgr)
        break
    prvs = next

cv2.destroyAllWindows()
```
